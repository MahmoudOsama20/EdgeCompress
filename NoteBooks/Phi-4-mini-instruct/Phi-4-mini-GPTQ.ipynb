```python
import os
import torch
import random
import numpy as np

from transformers import AutoTokenizer
from gptqmodel import GPTQModel, QuantizeConfig

from datasets import load_dataset
from huggingface_hub import login, upload_folder
from kaggle_secrets import UserSecretsClient
from huggingface_hub import create_repo
```

```python
# MODEL_ID_1 = "meta-llama/Llama-3.2-3B-Instruct"
MODEL_ID_2 = "microsoft/Phi-4-mini-instruct"
# OUTPUT_DIR_1 = "Llama-3.2-3B-Instruct-GPTQ-4bit"
OUTPUT_DIR_2 = "Phi-4-mini-instruct-GPTQ-4bit"

# REPO_ID_1 = "EdgeCompress01/Llama-3.2-3B-Instruct-GPTQ-4bit"
REPO_ID_2 = "EdgeCompress01/Phi-4-mini-instruct-GPTQ-4bit"

SEED = 42
```

```python
SEED = 42

def set_seed(seed):
    random.seed(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    torch.cuda.manual_seed_all(seed)

set_seed(SEED)
```

```python
def load_calibration(tokenizer):

    dataset = load_dataset(
        path="Salesforce/wikitext",
        name="wikitext-2-raw-v1",
        split="train"
    )

    dataset = dataset.shuffle(seed=42).select(range(256))

    calibration = []

    for text in dataset["text"]:

        text = text.strip()

        if len(text) == 0:
            continue

        tokens = tokenizer(
            text,
            truncation=True,
            max_length=128,
            return_tensors="pt"
        )

        calibration.append(tokens["input_ids"][0])

    return calibration
```

```python
def quantize_model(model_id, output_dir):

    print("Loading tokenizer...")

    tokenizer = AutoTokenizer.from_pretrained(
        model_id,
    )

    print("Creating GPTQ config...")

    quant_config = QuantizeConfig(
        bits=4,
        group_size=128,
        desc_act=False,
        sym=True,
        ignore_layers=[
            "lm_head",
            "embed_tokens"
        ]
    )

    print("Loading model...")

    model = GPTQModel.from_pretrained(
        model_id,
        quantize_config=quant_config,
        device="cuda"
    )

    examples = load_calibration(tokenizer)

    print("Running GPTQ quantization...")

    model.quantize(examples)

    print("Saving model...")

    model.save_pretrained(output_dir)
    tokenizer.save_pretrained(output_dir)

    return model, tokenizer
```

```python
qwen_model, tokenizer = quantize_model(
    MODEL_ID_2,
    OUTPUT_DIR_2
)
```

```python
def upload_model(folder_path, repo_id):

    create_repo(repo_id, repo_type="model", exist_ok=True)

    upload_folder(
        folder_path=folder_path,
        repo_id=repo_id,
        repo_type="model",
        commit_message="Upload GPTQ quantized model"
    )

upload_model(OUTPUT_DIR_2, REPO_ID_2)
```
